```python
img = cv2.cvtColor(img_1, cv2.COLOR_BGR2HSV)
figure_size = 7
result_image = cv2.blur(img, (figure_size, figure_size))

plt.figure(figsize=(11, 6))

plt.subplot(121), plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)), plt.title("Original")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_HSV2RGB)), plt.title("Blurred")
plt.xticks([]), plt.yticks([])

plt.show()
```

#Zooming an image

```python
(height, width) = img.shape[0:2]
zoom_img = cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(zoom_img)
```

```python
img_zoom = cv2.resize(img,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(img_zoom)
```

```python
img_r = cv2.imread("samples/butterfly.jpg")
plt.imshow(img_r)
```

#Rotating an Image

```python
height, width = img_r.shape[:2]

# Rotation matrix
img_rot = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)

# Apply rotation
rotate = cv2.warpAffine(img_r, img_rot, (width, height))

# Display the rotated image
plt.imshow(rotate)
plt.show()
```
